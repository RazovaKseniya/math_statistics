```python
def R2(x:np.ndarray, y:np.ndarray):
    
    ones = np.ones((x.shape[0],1))
    x = np.hstack([ones, x])
    
    XTX = x.T.dot(x)
    XTX_inv = np.linalg.inv(XTX)
    
    b = XTX_inv.dot(x.T).dot(y)
    
    z = x.dot(b)
    e = y - z
    
    return 1 - e.var() / y.var()

kd=[]
for i, j in combinations(range(X.shape[1]), 2):
    r = R2(X[:, [i,j]], y)
    print(i, j, r)
    kd.append(r)

print("Ответ - ",round(max(kd),8))
```

```python
d = 0.76342462
X = np.vstack((X[0:, 0], X[0:, 2])).T

n = X.shape[0]
m = X.shape[1]

k1 = m
k2 = n - m - 1

alpha = 0.05

t = stats.f.ppf(1 - alpha, k1, k2)

F = (d / k1) / ((1 - d) / (k2))
t, F
```

Статистика входит в критическую область. Ответ: 1
